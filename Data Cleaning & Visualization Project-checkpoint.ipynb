## Dataset Description
The Iris Dataset is a well-known dataset in data science used for classification and analysis.

Total Records: 150
Features:
Sepal Length
Sepal Width
Petal Length
Petal Width
Species (Setosa, Versicolor, Virginica)
Objective:

To clean, analyze, and visualize the dataset to extract meaningful insights about flower species.

“This notebook serves as a visual dashboard summarizing key insights from the Iris dataset.”

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## load dataset
df = pd.read_csv("Iris.csv")
df.head()
```

```python
## check missing values
df.isnull().sum()
```

```python

print(df.columns)
```

```python
## check duplicates
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
## check data types
df.info()
```

```python
## removed duplicates
df.duplicated().sum()
```

```python
## detect outliers
sns.boxplot(data=df)
plt.show()
```

## “Outliers were observed in the Sepal Width feature, while other features showed relatively normal distribution without significant outliers.”

```python
## convert species to category
df['Species'] = df['Species'].astype('category')
```

```python
df.info()
```

### Data Analysis

```python
df.describe()
```

```python
df['Species'].value_counts()
```

## data visualization

```python
## Pairplot (Best Visualization)
sns.pairplot(df, hue="Species")
plt.show()
```

## Shows clear separation between species.

```python
## histogram
df.hist(figsize=(10,8))
plt.show()
```

```python
## boxplot (species comparision)
sns.boxplot(x="Species", y="PetalLengthCm", data=df)
plt.show()
```

# “A boxplot of petal length across species shows clear differentiation among the three species. Iris-setosa has the smallest petal length, while Iris-virginica has the largest. Minor outliers were observed in Iris-setosa and Iris-versicolor, but they do not significantly impact the overall distribution.”

```python
## scatterplot
sns.scatterplot(x="SepalLengthCm", y="PetalLengthCm", hue="Species", data=df)
plt.show()
```

# “The scatter plot between sepal length and petal length shows a positive correlation. Iris-setosa forms a clearly distinct cluster with low values, while Iris-versicolor and Iris-virginica show partial overlap. However, Iris-virginica generally exhibits higher values, making petal length a strong distinguishing feature.” 

```python
## heatmap (correlation)
numeric_df = df.select_dtypes(include=['float64', 'int64'])
sns.heatmap(numeric_df.corr(), annot=True)
plt.show()
```

## From the correlation matrix, it can be observed that Petal Length and Petal Width have a strong positive correlation, while Sepal Width shows weak or negative correlation with the other features. This dataset is useful for understanding feature relationships and building predictive models.

```python
## dashboard
import plotly.express as px

fig = px.scatter(df, 
                 x="SepalLengthCm", 
                 y="PetalLengthCm", 
                 color="Species",
                 title="Iris Dashboard Scatter")

fig.show()
```

## The visualization clearly demonstrates the clustering of different species. For instance, Iris-setosa (purple) forms a distinct, low-value cluster, while Iris-versicolor (orange) and Iris-virginica (green) show a linear correlation between sepal and petal length, though with specific areas of overlap that the dashboard helps identify.
